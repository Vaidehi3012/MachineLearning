1. What is Dimensionality Reduction? Why is it important in machine
learning?

  - Dimensionality Reduction: A Comprehensive Explanation
1. Definition:

  - Dimensionality reduction is a set of techniques used in machine learning and statistics to reduce the number of random variables under consideration by obtaining a set of principal variables. It's the process of transforming data from a high-dimensional space into a lower-dimensional space such that the lower-dimensional representation retains some meaningful properties of the original data, ideally close to the intrinsic dimension of the data. This transformation can be either a feature selection (selecting a subset of the original features) or feature extraction (transforming the data into a new set of features).

2. Rationale and Motivation:

  - The primary motivation for dimensionality reduction stems from the challenges posed by high-dimensional data, often referred to as the "Curse of Dimensionality."

  - The Curse of Dimensionality: As the number of features or dimensions in a dataset increases, the volume of the space increases so fast that the available data becomes sparse. This sparsity makes it difficult to find statistically significant relationships, leads to unreliable distance calculations, and increases the computational burden for many algorithms.
3. Importance and Benefits in Machine Learning:

  - Dimensionality reduction offers several critical advantages in the context of machine learning:

  - Mitigation of the Curse of Dimensionality: It directly addresses the problems caused by high-dimensional data, preventing overfitting and improving the generalization capabilities of models by reducing noise and redundancy.
  - Improved Model Performance and Accuracy: By removing irrelevant and redundant features, models can focus on the most informative aspects of the data. This often leads to simpler models that are less prone to overfitting and can achieve higher predictive accuracy.
  - Faster Training and Reduced Computational Cost: Fewer features mean less data to process. This significantly reduces the training time for machine learning algorithms and lowers the computational resources required for both training and inference. For very large datasets, this can be crucial for practical deployment.
  - Reduced Storage Requirements: Storing data with fewer dimensions requires less memory and disk space, which can be beneficial for large datasets or resource-constrained environments.
  - Enhanced Data Visualization: It's practically impossible to visualize data in more than three dimensions. Dimensionality reduction techniques allow complex, high-dimensional data to be projected into two or three dimensions, enabling human interpretation, pattern recognition, and exploratory data analysis.
  - Noise Reduction and Feature Engineering: Often, high-dimensional datasets contain noisy features that do not contribute to the underlying patterns but rather obscure them. Dimensionality reduction can act as a form of noise filtering by focusing on the components that explain the most variance.
  - Better Interpretability: A reduced set of features can sometimes be more interpretable, especially when the extracted features represent underlying latent factors.
4. Common Techniques:

Dimensionality reduction techniques are broadly categorized into two main types:

  - Feature Selection: Involves selecting a subset of the most relevant original features. Methods include Filter methods (e.g., variance threshold, correlation), Wrapper methods (e.g., recursive feature elimination), and Embedded methods (e.g., Lasso regression).
  - Feature Extraction: Transforms the data from the high-dimensional space to a space with fewer dimensions. This new, smaller set of features is a combination of the original features.
  - Principal Component Analysis (PCA): A linear technique that identifies the principal components (new orthogonal features) that capture the maximum variance in the data.
  - Linear Discriminant Analysis (LDA): A supervised technique used for classification that finds a linear combination of features that separates two or more classes of objects or events.
  - t-Distributed Stochastic Neighbor Embedding (t-SNE): A non-linear technique particularly well-suited for visualizing high-dimensional datasets in 2D or 3D, preserving local structures.
  - Uniform Manifold Approximation and Projection (UMAP): Another non-linear technique, often faster than t-SNE, that aims to preserve both local and global data structures.
5. Challenges and Considerations:

  - Loss of Information: By reducing dimensions, some information is invariably lost. The goal is to minimize the loss of relevant information while discarding noise.
  - Interpretability of New Features: Especially with feature extraction methods like PCA, the new components might not have a direct, intuitive meaning, making interpretation challenging.
  - Choosing the Right Technique and Number of Dimensions: The effectiveness of dimensionality reduction heavily depends on the technique chosen and the optimal number of dimensions to retain, which often requires experimentation and domain knowledge.

2. Name and briefly describe three common dimensionality reduction
techniques.

  - 1. Principal Component Analysis (PCA)

  - Description: PCA is a linear dimensionality reduction technique that transforms a dataset into a new coordinate system such that the greatest variance by any projection of the data comes to lie on the first coordinate (called the first principal component), the second greatest variance on the second coordinate, and so on. Essentially, it finds the directions (principal components) along which the data varies the most.
  - How it Works:
  - Standardization: Data is typically standardized to have zero mean and unit variance.
  - Covariance Matrix Calculation: The covariance matrix of the data is computed to understand the relationships between different features.
  - Eigenvalue Decomposition: Eigenvalues and eigenvectors of the covariance matrix are calculated. Eigenvectors represent the principal components (directions of maximum variance), and eigenvalues represent the magnitude of variance along those directions.
  - Component Selection: Principal components are ordered by their corresponding eigenvalues in descending order. A subset of these components with the largest eigenvalues is selected to form the new, lower-dimensional feature space, retaining most of the data's variance.
  - Characteristics:
  - Unsupervised: It does not use class labels (target variable) to perform dimensionality reduction.
  - Linear: It identifies linear combinations of the original features.
  - Variance Maximization: Focuses on preserving as much variance as possible.
  - Orthogonal Components: The principal components are uncorrelated (orthogonal).
  - Use Cases: Data compression, noise reduction, feature extraction, and exploratory data analysis.
2. Linear Discriminant Analysis (LDA)

  - Description: LDA is a supervised dimensionality reduction technique primarily used for classification problems. Unlike PCA, which focuses on maximizing variance, LDA aims to find a linear combination of features that best separates different classes in the dataset. It projects the data onto a lower-dimensional space such that the inter-class variance is maximized, and the intra-class variance is minimized.
  - How it Works:
  - Calculate Class Means: Determine the mean vector for each class.
  - Calculate Scatter Matrices: Compute the within-class scatter matrix (measures variance within each class) and the between-class scatter matrix (measures variance between class means).
  - Eigenvalue Decomposition: Solve the generalized eigenvalue problem for the scatter matrices to find the eigenvectors (linear discriminants) that maximize the ratio of between-class scatter to within-class scatter.
  - Component Selection: Select the eigenvectors corresponding to the largest eigenvalues to form the new, lower-dimensional space.
  - Characteristics:
  - Supervised: Requires class labels to perform dimensionality reduction.
  - Linear: Identifies linear combinations of features.
  - Discriminative: Focuses on maximizing class separability.
  - Maximum Components: Can reduce dimensionality to at most C-1 dimensions, where C is the number of classes.
  - Use Cases: Facial recognition, medical diagnosis, customer segmentation for classification, and as a preprocessing step for classification algorithms.
3. t-Distributed Stochastic Neighbor Embedding (t-SNE)

  - Description: t-SNE is a non-linear dimensionality reduction technique particularly well-suited for visualizing high-dimensional datasets in 2D or 3D. It aims to preserve local structures in the data, meaning that points that are close together in the high-dimensional space remain close together in the low-dimensional embedding.
  - How it Works:
  - High-Dimensional Probability Distribution: Constructs a probability distribution over pairs of high-dimensional data points, where similar points have a higher probability of being neighbors.
  - Low-Dimensional Probability Distribution: Creates a similar probability distribution in the lower-dimensional space (typically 2D or 3D).
  - Kullback-Leibler Divergence Minimization: Uses a gradient descent optimization to minimize the Kullback-Leibler (KL) divergence between these two probability distributions. This process iteratively adjusts the positions of points in the low-dimensional map until the two distributions are as similar as possible.
  - Student's t-distribution: Employs a Student's t-distribution in the low-dimensional space to alleviate the "crowding problem," which allows dissimilar points to be modeled far apart.
  - Characteristics:
  - Non-linear: Can capture complex, non-linear relationships in the data.
  - Unsupervised: Does not require class labels, though labels can be used for coloring the resulting visualization.
  - Emphasis on Local Structure: Excels at preserving local neighborhoods, making it ideal for visualizing clusters.
  - Computational Cost: Can be computationally intensive for very large datasets.
  - Use Cases: Data exploration, visualization of complex datasets (e.g., genomic data, image features, text embeddings), and identifying clusters or manifold structures.

3. What is clustering in unsupervised learning? Mention three popular
clustering algorithms

  - Clustering is a core technique within unsupervised learning, a branch of machine learning where the algorithm learns from data that has not been labeled or categorized. The primary objective of clustering is to discover inherent groupings or structures within a dataset without any prior knowledge of those groupings.

  - More formally, clustering aims to partition a set of n data points into k distinct subsets, or clusters, such that:

  - High Intra-cluster Similarity: Data points within the same cluster are highly similar to each other based on a chosen similarity or distance metric (e.g., Euclidean distance, cosine similarity).
  - Low Inter-cluster Similarity: Data points belonging to different clusters are dissimilar to each other.
  - The algorithm identifies these patterns and groups based purely on the features of the data itself. Clustering is an exploratory data analysis tool, often used to understand the underlying distribution of data, segment populations, detect anomalies, or serve as a preprocessing step for other supervised learning tasks.

  - Three Popular Clustering Algorithms:
1. K-Means Clustering

  - Description: K-Means is one of the most widely used and well-known centroid-based clustering algorithms. It is an iterative algorithm that aims to partition n observations into k pre-defined clusters, where each observation belongs to the cluster whose centroid (mean) is closest.
  - How it Works:
  - Initialization: The algorithm begins by randomly selecting k data points from the dataset as the initial cluster centroids.
  - Assignment Step: Each data point is assigned to the cluster whose centroid it is closest to (typically using Euclidean distance).
  - Update Step: After all data points have been assigned, the centroids of the k clusters are recalculated. The new centroid for each cluster is the mean of all data points currently assigned to that cluster.
  - Iteration: Steps 2 and 3 are repeated iteratively until a convergence criterion is met. This usually happens when the cluster assignments no longer change, or the movement of centroids falls below a certain threshold, or a maximum number of iterations is reached.
  - Characteristics:
  - Simple and Efficient: Computationally fast for large datasets.
  - Requires k: The number of clusters (k) must be specified by the user beforehand.
  - Sensitive to Initialization: The final clustering result can depend on the initial random placement of centroids; often run multiple times with different initializations.
  - Assumes Spherical Clusters: Works best with clusters that are roughly spherical and of similar size and density.
  - Sensitive to Outliers: Outliers can significantly pull centroids, distorting cluster shapes.
  - Use Cases: Customer segmentation, image compression, document clustering, anomaly detection (by identifying small, isolated clusters).
2. Hierarchical Clustering (Agglomerative)

  - Description: Hierarchical clustering builds a hierarchy of clusters, represented as a tree-like diagram called a dendrogram. Agglomerative (bottom-up) hierarchical clustering is the most common type. It starts by treating each data point as an individual cluster and then successively merges clusters until all points are in a single cluster or a desired number of clusters is achieved.
  - How it Works (Agglomerative):
  - Initialization: Each data point is considered a single, individual cluster.
  - Distance Calculation: The distances (or dissimilarities) between all pairs of clusters are calculated. A 'linkage criterion' defines how the distance between two clusters is measured (e.g., single linkage: minimum distance between any two points in the clusters; complete linkage: maximum distance; average linkage: average distance; Ward's method: minimizes variance within merged clusters).
  - Merging: The two closest clusters (based on the chosen linkage criterion) are merged into a new, larger cluster.
  - Iteration: Steps 2 and 3 are repeated until all data points are grouped into a single, comprehensive cluster.
  - Dendrogram: The entire process generates a dendrogram, which can be 'cut' at a certain height to obtain a desired number of clusters.
  - Characteristics:
  - No Pre-defined k: Does not require specifying the number of clusters beforehand; k can be chosen by inspecting the dendrogram.
  - Visual Representation: The dendrogram provides a rich visualization of the cluster structure and relationships.
  - Computationally Intensive: Can be slow for large datasets due to the need to calculate and update distance matrices.
  - Irreversible: Once a merge is made, it cannot be undone.
  - Use Cases: Biological classification (taxonomy), social network analysis, market research, and any domain where understanding the hierarchical relationships between data points is important.
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

  - Description: DBSCAN is a density-based clustering algorithm that groups together data points that are closely packed together (i.e., having a high density of points) and marks as outliers (noise) points that lie alone in low-density regions. It can discover clusters of arbitrary shapes and is robust to outliers.
  - How it Works:
  - Parameters: DBSCAN requires two parameters: eps (Îµ), which defines the maximum radius of the neighborhood to consider for a point, and MinPts, the minimum number of points required to form a dense region.
  - Core Points: A data point p is classified as a core point if at least MinPts (including p itself) are found within its eps neighborhood.
  - Border Points: A data point q is a border point if it is not a core point but lies within the eps neighborhood of a core point.
  - Noise Points: All other data points that are neither core nor border points are classified as noise points (outliers).
  - Cluster Formation: The algorithm starts with an arbitrary unvisited data point. If it's a core point, a new cluster is started, and all density-reachable points (directly reachable core points and their border points) are added to this cluster. This process continues until no more points can be added to the current cluster. The algorithm then moves to the next unvisited point.
  - Characteristics:
  - Arbitrary Cluster Shapes: Can find clusters of complex, non-linear shapes, unlike K-Means.
  - Handles Noise: Effectively identifies and isolates outliers as noise points.
  - No Pre-defined k: Does not require specifying the number of clusters beforehand; it discovers them based on density.
  - Sensitive to Parameters: The results can be highly sensitive to the eps and MinPts parameters, which often require domain knowledge or experimentation to tune.
  - Struggles with Varying Densities: May have difficulty separating clusters of widely varying densities.
  - Use Cases: Anomaly detection (outlier identification), spatial data mining, geology, and any application where clusters may be non-spherical or noise is prevalent.

4. Explain the concept of anomaly detection and its significance.

  - Anomaly Detection: Concept and Significance
  - Concept of Anomaly Detection:

  - Anomaly detection, also known as outlier detection, is the process of identifying data points, events, or observations that deviate significantly from the majority of the data. These "anomalies" are often referred to as outliers, novelties, noise, deviations, or exceptions. They represent unusual patterns that do not conform to expected behavior.

  - Anomalies can manifest in various forms:

  - Point Anomalies: A single data instance is anomalous with respect to the rest of the data. (e.g., an unusually high credit card transaction for a user).
  - Contextual Anomalies: A data instance is anomalous in a specific context, but not otherwise. (e.g., a high temperature reading is normal in summer but anomalous in winter).
  - Collective Anomalies: A collection of related data instances is anomalous, even if individual instances are not. (e.g., a series of small but coordinated network access attempts).
  - The goal of anomaly detection is to build a model of "normal" behavior and then identify any observations that do not fit this model. This often involves statistical methods, machine learning algorithms, or domain-specific rules.

  - Significance of Anomaly Detection:

  - Anomaly detection is of paramount importance across a wide range of fields due to its ability to uncover critical, often rare, events that can have significant implications. Its significance stems from several key benefits:

  - Fraud Detection: In finance, banking, and e-commerce, anomalies can indicate fraudulent transactions, unusual spending patterns, or account compromises. Detecting these quickly can prevent substantial financial losses.
  - Cybersecurity: Identifying unusual network traffic patterns, suspicious login attempts, or abnormal system behavior can signal cyberattacks, intrusions, or malware infections, protecting sensitive data and infrastructure.
  - System Health Monitoring: In IT operations and industrial control systems, anomalies in sensor readings (e.g., temperature, pressure, vibration) can indicate equipment malfunctions, impending failures, or operational inefficiencies, allowing for proactive maintenance and preventing costly downtime.
  - Medical Diagnosis: Deviations from normal physiological parameters (e.g., heart rate, blood pressure, MRI scans) can be indicators of diseases or medical conditions, aiding in early diagnosis and treatment.
  - Quality Control: In manufacturing, anomalies in product characteristics can signal defects or process errors, helping maintain product quality and reduce waste.
  - Scientific Discovery: In scientific research, outliers can sometimes represent novel phenomena, unexpected experimental results, or errors in data collection that, upon investigation, lead to new insights or discoveries.
  - Environmental Monitoring: Unusual readings from environmental sensors can indicate pollution incidents, climate change effects, or natural disasters, prompting timely interventions.
  - Customer Behavior Analysis: Identifying anomalous customer behavior can reveal unusual purchasing patterns, potential churn risks, or even opportunities for targeted marketing if the anomaly represents a new trend.

5. List and briefly describe three types of anomaly detection techniques.

  - 1. Statistical-Based Techniques:

  - Description: These methods assume that normal data follows a specific statistical distribution (e.g., Gaussian). Anomalies are identified as data points that fall into low-probability regions of this distribution, deviating significantly from the expected statistical properties.
  Example: Using Z-scores to flag data points that are many standard deviations away from the mean.
2. Machine Learning-Based Techniques:

  Description: These techniques leverage various machine learning algorithms to learn patterns of normal behavior from data. Anomalies are then identified as instances that do not conform to these learned patterns. These can be unsupervised (most common, e.g., Isolation Forest, One-Class SVM), supervised (if labeled data is available), or semi-supervised (trained on normal data only).
  Example: An Isolation Forest isolates anomalies by randomly selecting features and then randomly selecting a split value between the maximum and minimum values of the selected feature. Anomalies are points that require fewer splits to be isolated.
3. Proximity-Based Techniques:

  Description: These methods define anomalies based on their distance or density relative to other data points. The core idea is that normal data points are usually found in dense neighborhoods, close to many other points, while anomalies are isolated or lie far from their neighbors.
  Example: Local Outlier Factor (LOF) quantifies the local density deviation of a data point with respect to its neighbors. Points with significantly lower density than their neighbors are considered outliers.

6. What is time series analysis? Mention two key components of time series
data.

  - Time series analysis is a statistical technique that deals with time series data, which are sequences of data points indexed (or listed or graphed) in time order. This type of data is collected over successive time intervals, meaning each data point corresponds to a specific point in time.

  - The primary goal of time series analysis is to extract meaningful statistics and other characteristics from the data, which can then be used for forecasting (predicting future values), understanding underlying causes, or detecting patterns. It involves methods for analyzing time series data in order to extract meaningful statistics and other characteristics of the data.

  - In essence, time series analysis aims to:

  - Understand the past: Identify patterns, trends, and cyclical behaviors within the historical data.
  - Explain the present: Analyze how different factors or events influence the current state of the time series.
  - Predict the future: Develop models to forecast future values of the series, which is crucial for planning, resource allocation, and decision-making.
  - Detect anomalies: Identify unusual data points or patterns that deviate significantly from expected behavior over time.
  - Time series data differs from regular cross-sectional data because the order of observations is crucial. The dependencies between consecutive observations are a central aspect of its analysis.

  - Two Key Components of Time Series Data:
  - Time series data can often be decomposed into several components that help in understanding and modeling its behavior. Two of the most fundamental and widely recognized components are:

1. Trend Component (T_t):

  - Description: The trend component represents the long-term upward or downward movement in the data over an extended period. It reflects the underlying growth or decline in the series, often due to fundamental changes like population growth, technological advancements, economic development, or shifting consumer preferences. A trend can be linear or non-linear, and it does not necessarily follow a straight line.
  - Characteristics: It captures the general direction of the series. For example, a country's GDP might show an upward trend over decades, or the sales of a product might show a downward trend as it approaches obsolescence. It is a smooth, slowly changing component.
  - Importance: Identifying the trend is crucial for understanding the overall direction and stability of the series. It helps in making long-term forecasts and understanding macro-level changes affecting the variable.
2. Seasonal Component (S_t):

  - Description: The seasonal component refers to repetitive, predictable patterns or fluctuations that occur within a fixed period, such as a year, month, week, or even a day. These patterns are typically caused by factors that repeat annually (e.g., holidays, weather changes), monthly (e.g., billing cycles), or weekly (e.g., weekend sales).
  - Characteristics: Seasonal variations have a fixed and known period. For example, ice cream sales typically peak in summer and dip in winter; retail sales often surge during the Christmas season; daily electricity consumption might peak during business hours. This component is regular and predictable.
  - Importance: Recognizing and quantifying seasonality is vital for short-to-medium term forecasting and for making informed operational decisions. By removing the seasonal component (seasonal adjustment), analysts can better observe the underlying trend and cyclical movements.

7. Describe the difference between seasonality and cyclic behavior in time
series.

  - Seasonality in Time Series
  - Definition: Seasonality refers to predictable and regular patterns that repeat over a fixed and known period. These patterns are typically tied to calendar divisions, such as daily, weekly, monthly, quarterly, or yearly cycles.
  - Cause: Driven by factors like time of day, day of the week, month of the year, public holidays, or weather patterns.
  - Periodicity: The length of the cycle is constant and known. For example, daily patterns repeat every 24 hours, weekly patterns every 7 days, and yearly patterns every 12 months.
  - Predictability: Highly predictable in terms of timing and amplitude (though the exact values may vary).
  - Examples:
  - Retail sales peaking every December due to Christmas.
  - Electricity consumption rising every afternoon.
  - Ice cream sales increasing every summer.

  - Cyclic Behavior in Time Series
  - Definition: Cyclic behavior refers to fluctuations in a time series that are not of fixed period. These cycles can last for several years and are often associated with economic or business cycles.
  - Cause: Often influenced by broader economic factors, business expansion and contraction, market trends, or other long-term phenomena.
  - Periodicity: The length of a cycle is variable and unknown. Cycles are irregular in their duration and amplitude, making them harder to predict than seasonal patterns.
  - Predictability: Less predictable than seasonality, as their duration and magnitude can vary significantly.
  - Examples:
  - Boom and bust cycles in the economy.
  - Periods of high and low unemployment over several years.
  - Long-term trends in commodity prices.

  - Feature	Seasonality	Cyclic Behavior
Periodicity	Fixed and Known (e.g., daily, weekly, yearly)	Variable and Unknown (often several years)

Length	Relatively short (up to a year)	Long-term (more than a year)

Cause	Calendar-related, specific events	Economic conditions, business cycles, broad trends

Predictability	High	Low

Regularity	Regular, consistent patterns	Irregular, varying duration and amplitude

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Generate a sample dataset
# We'll create a dataset with 3 distinct 'blobs' or clusters for demonstration
n_samples = 300
n_features = 2 # 2D data for easy visualization
n_clusters = 3 # We'll assume we know there are 3 clusters for K-Means

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters,
                  cluster_std=0.60, random_state=0)

print(f"Shape of the dataset: {X.shape}")

# 2. Visualize the original (unclustered) data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title('Original (Unclustered) Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# 3. Perform K-means clustering
# Initialize KMeans with the number of clusters (n_clusters)
# random_state ensures reproducibility
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init=10)

# Fit the model to the data and predict the clusters
y_kmeans = kmeans.fit_predict(X)

# Get the cluster centers
centers = kmeans.cluster_centers_

# 4. Visualize the clustered data
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7, marker='X', label='Centroids')
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.tight_layout()
plt.show()
```

9. What is inheritance in OOP? Provide a simple example in Python.

  - Inheritance in Object-Oriented Programming (OOP) is a mechanism that allows a new class (called the child class or subclass) to derive properties and behaviors (attributes and methods) from an existing class (called the parent class or superclass).

  - It promotes code reusability, reduces redundancy, and establishes an 'is-a' relationship between classes, allowing child classes to extend or specialize the functionality of their parents.

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"Brand: {self.brand}, Model: {self.model}"

    def start_engine(self):
        return "Engine started!"


class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)  # Call the parent class's constructor
        self.doors = doors

    def drive(self):
        return f"The {self.brand} {self.model} is driving with {self.doors} doors."


class Motorcycle(Vehicle):
    def __init__(self, brand, model, type_of_bike):
        super().__init__(brand, model)  # Call the parent class's constructor
        self.type_of_bike = type_of_bike

    def ride(self):
        return f"The {self.brand} {self.model} ({self.type_of_bike}) is being ridden."


# Create objects of the child classes
my_car = Car("Toyota", "Camry", 4)
my_motorcycle = Motorcycle("Harley-Davidson", "Fat Boy", "Cruiser")

# Access methods from both parent and child classes
print("--- Car Information ---")
print(my_car.display_info())     # Inherited method
print(my_car.start_engine())    # Inherited method
print(my_car.drive())           # Child class's specific method

print("\n--- Motorcycle Information ---")
print(my_motorcycle.display_info())  # Inherited method
print(my_motorcycle.start_engine()) # Inherited method
print(my_motorcycle.ride())         # Child class's specific method
```

10. How can time series analysis be used for anomaly detection?

  - Time series analysis is a powerful tool for anomaly detection because it leverages the sequential nature and inherent patterns (like trend and seasonality) of data over time to identify deviations from expected behavior. Here's how it's typically used:

  - Modeling Normal Behavior: The core idea is to build a model that accurately represents the 'normal' patterns within a time series. This model learns the expected values, trends, and seasonal fluctuations. Techniques like ARIMA (AutoRegressive Integrated Moving Average), Prophet, or even simpler statistical methods can be used to capture these patterns.

  - Predicting Future Values: Once a model of normal behavior is established, it can be used to predict the next expected value (or range of values) in the time series.

  - Comparing Predictions to Actuals: The predicted values are then compared against the actual observed values. Any significant difference between the actual observation and the model's prediction suggests a potential anomaly.

  - Defining Anomaly Thresholds: A threshold is set to determine what constitutes a 'significant difference'. This could be based on statistical measures (e.g., standard deviations from the prediction, prediction intervals), domain knowledge, or machine learning-based methods (e.g., classifying residuals as anomalous or not).

  - Specific approaches and considerations in time series anomaly detection:

  - Residual Analysis: After fitting a time series model (e.g., ARIMA), the residuals (the differences between observed and predicted values) are analyzed. Large, unusual residuals are strong indicators of anomalies. Standard statistical anomaly detection techniques can then be applied to these residuals.
  - Forecasting-Based Methods: These involve training a forecasting model on historical data. When a new data point arrives, the model predicts its value, and if the actual value falls outside a pre-defined prediction interval (e.g., 95% confidence interval), it's flagged as an anomaly.
  - Statistical Control Charts: Commonly used in quality control, these charts track a time series and flag points that fall outside control limits, indicating a process is out of statistical control.
  - Machine Learning for Time Series: Algorithms like Isolation Forest, One-Class SVM, or even deep learning models (e.g., LSTMs, Autoencoders) can be adapted for time series anomaly detection. They learn temporal dependencies and flag observations that don't conform to the learned sequence of normal events.
  - Autoencoders: Can be trained to reconstruct normal time series sequences. Anomalies will result in high reconstruction errors.
  - Recurrent Neural Networks (RNNs) / LSTMs: Can learn complex temporal patterns. Anomalies are detected when the network predicts a value far from the actual, or when the hidden state deviates significantly.
  - Change Point Detection: While related, anomaly detection often focuses on single unusual data points, whereas change point detection looks for shifts in the underlying statistical properties of the time series itself (e.g., a sudden change in mean or variance).
